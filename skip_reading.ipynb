```python
import pandas as pd
```

Function writen to skip through the entire database and select a portion. This is faster and more memory efficient that random sampling; however, this does not give an accurate representation of the data. This is was done to get a crude and broad representation of the data since it does not fit into memory.

```python
file_path = "/app/data/bond_56_million_enriched.csv"
```

```python
def SkipReader(file_path,n_rows_to_skip, n_rows_to_read, dtypes, col_names):
    #function to skip through csv and read
    
    temp_df = pd.read_csv(filepath_or_buffer=file_path,
                          header=0,
                          nrows=n_rows_to_read,
                          skiprows=n_rows_to_skip,
                          low_memory=False,
                          dtype=dtypes,
                          names=col_names)
    
    return temp_df
```

```python
n_rows_to_read = 1000000
n_rows_to_skip = 6000000
chunk = n_rows_to_read + n_rows_to_skip

df = pd.read_csv(file_path, header=0, nrows=1000000, low_memory=False)

dtypes_dict = df.dtypes.apply(lambda x: x.name).to_dict()
column_names = list(df.columns)

for i in range(8):
    n_rows_to_skip = chunk * i
    
    df_to_concat = SkipReader(file_path,
                              n_rows_to_skip=n_rows_to_skip,
                              n_rows_to_read=n_rows_to_read,
                              dtypes=dtypes_dict,
                              col_names=column_names)
    
    df = pd.concat([df, df_to_concat])    
```
